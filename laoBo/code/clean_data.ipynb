```python
#数据可视化
import numpy as np
import matplotlib.pyplot as plt

#运动员男女比例
ath_sex = data_ath_c["Sex"]
size_sex = ath_sex.value_counts(normalize = True).reset_index()
size_sex.columns = ["sex","size"]


plt.figure(figsize=(8, 8))  # 设置图形的大小
plt.pie(size_sex["size"], labels=size_sex["sex"], autopct='%1.1f%%', startangle=140)
plt.title('')
plt.axis('equal')  # 保证饼图是一个圆形
plt.show()
```

```python
#数据统计
#统计国家获得奖牌
#考虑到每年的奖牌总数不断变化，我们对奖牌数进行标准化
data_med = data_med_c.copy()
med_year = data_med.groupby("Year")   #按照年份进行划分
data_med["year_total"] = med_year["Total"].transform("sum")
data_med["Gold_size"] = data_med["Gold"] / data_med["year_total"]
data_med["Silver_size"] = data_med["Silver"] / data_med["year_total"]
data_med["Bronze_size"] = data_med["Bronze"] / data_med["year_total"]
data_med["Total_size"] = data_med["Total"] / data_med["year_total"]

med_country = data_med.groupby("NOC")[["Gold_size","Silver_size","Bronze_size","Total_size"]].mean().reset_index()
med_country = med_country.rename(columns = {"NOC":"country"})
```

```python
#各国历史运动员数量
#因为国家数量过多，所以只取前100个
import seaborn as sns
ath_country = data_ath_c["Country"]
size_country = ath_country.value_counts(normalize = False).reset_index()
size_country.columns= ["country","size"]

#对应的奖牌比例
size_med = pd.merge(size_country,med_country,on = "country",how ="left")
size_med = size_med[:50]
    
sns.set(style= "whitegrid")
# 绘制条线图
plt.figure(figsize=(14, 8))  # 设置图形大小
ax1 = sns.barplot(x = "country", y = "size", data = size_med, color = "Gold", palette='pastel')
ax1.set_xlabel("NOC")
ax1.set_ylabel("num of ath")
ax1.tick_params(axis='y', labelcolor='blue')
plt.xticks(rotation=90)



ax2 = ax1.twinx()
#绘制折线图
sns.lineplot(data=size_med,
              x='country', 
              y='Gold_size', 
              label = "Gold" ,
              marker='o')
sns.lineplot(data=size_med,
              x='country', 
              y='Silver_size',  
              label = "Silver",
              marker='o')
sns.lineplot(data=size_med,
              x='country', 
              y='Bronze_size', 
              label = "Bronze", 
              marker='o')
sns.lineplot(data=size_med,
              x='country', 
              y='Total_size',  
              label = "Total",
              marker='o')
ax2.set_ylabel("propotion of medal",color = "black")
ax2.tick_params(axis='y', labelcolor='black')

# 添加标题和标签
plt.title('medal and athletes', fontsize=16)
plt.xlabel('country', fontsize=14)
plt.ylabel('mean propotion', fontsize=14)

# 添加网格
plt.grid(True, linestyle='--', alpha=0.5)


# 显示图形
plt.show()
```

```python
col = ["Sport","Code"] + list(data_pro_c.columns[4:])
data_sport = data_pro_c.loc[:,col]

#缺失值处理
#s5:因为项目取消，直接用0填充
#.:展示项目，直接用0代替
#部分项目没有code，用缩写代替
data_sport.fillna(0,inplace = True)
data_sport = data_sport.replace("•",0)
data_sport.loc[39,"Code"] = "JDP"
data_sport.loc[47,"Code"] = "RQE"
data_sport.loc[69,"1924"] = 0
data_sport.loc[70,"1924"] = 0
```

```python
from matplotlib.colors import ListedColormap

#统计项目变化
num = []
for i in range(31):
    num.append(i)

plt.figure(figsize=(20, 15))

cmap = sns.color_palette("viridis", as_cmap=True)
new_cmap = ListedColormap(['white'] + cmap.colors)

# 创建一个遮罩，零值和NaN值为True，其余为False
mask = (data_sport.isnull()) | (data_sport == 0)

# 绘制热力图，屏蔽零值和NaN值
sns.heatmap(data_sport, 
            mask=mask, cmap='viridis', 
            linewidths=0.5, 
            linecolor='gray', 
            cbar=True, 
            cbar_kws={'label': '数量'}, 
            square=True,
            
            annot_kws={"size": 10})

# 重新绘制零值部分，使用白色或其他明显颜色
sns.heatmap(data_sport, 
            mask=~mask, 
            cmap=ListedColormap(['white']),
            linewidths=0.5, 
            linecolor='gray', 
            cbar=False, 
            alpha=0.3
           )

plt.title('各运动在奥林匹克的数量变化热力图', fontsize=18)
plt.xlabel('年份', fontsize=14)
plt.ylabel('运动名称', fontsize=14)
plt.show()
```
